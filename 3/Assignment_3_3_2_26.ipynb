Assignment 3: Visualize the data using Python by plotting the graphs for assignment no. 1 and 2. Consider a suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

noisy_data = {
    'StudentID': [101, 102, 103, 104, 105, 106, 107, 108,
                  109, 110, 111, 112, 113, 114, 115, 115],

    'Name': ['Aarav', 'Diya', 'Rohan', 'Sneha', 'Karan', 'Isha',
             'Aryan', 'Pooja', 'Rahul', 'Neha', 'Aman', 'Priya',
             'Sahil', 'Ananya', 'Vikram', 'Adil'],

    'Age': [19, 20, 18, 21, 21, 19, 22, 20,
            20, 19, 21, 22, 19, 18, 23, 23],

    'Gender': ['M', 'F', 'M', 'F', 'M', 'F',
               'M', 'F', 'M', 'F', 'M', 'F',
               'M', 'F', 'M', 'M'],

    'Department': ['CSE', 'IT', 'CSE', 'ECE', 'ME', 'CSE',
                   'IT', 'ECE ', 'ME', 'CSE', 'IT', 'IT',
                   'CSE', 'ME', 'IT', 'IT'],

    'Marks': [85, 91, 78, 87, 72, 95,
              90, 84, 77, 92, 76, 89,
              83, 68, 90, 91],

    'Attendance': [92, 95, 88, 90, 91, 98,
                   89, 65, 80, 96, 70, 93,
                   90, 82, 94, 94],

    'Internship': ['Yes', 'Yes', 'No', 'Yes', 'No', 'Yes',
                   'No', 'Yes', 'No', 'Yes', 'No', 'Yes',
                   'Yes', 'No', 'Yes', 'Yes']
}

df = pd.DataFrame(noisy_data)
print(df)
```

Scatter Plot to find Marks v Attendance distribution

```python
import matplotlib.pyplot as plt

plt.scatter(df['Marks'], df['Attendance'])
plt.xlabel('Marks')
plt.ylabel('Attendance')
plt.title('Scatter Plot: Marks vs Attendance')
plt.show()
```

Bar Plot to weigh marks of students against each other


```python
plt.bar(df['Name'], df['Marks'], color='skyblue')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Bar Plot: Marks by Student')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

Box Plot of Marks Distribution

```python
plt.boxplot([df['Marks']], labels=['Marks'])
plt.ylabel('Marks')
plt.title('Box Plot: Marks Distribution')
plt.show()
```

Pie Chart of Percentage of students per department

```python
dept_counts = df['Department'].value_counts()
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart: Students by Department')
plt.show()
```

Line Chart of Marks Trend

```python
plt.plot(df['StudentID'], df['Marks'], marker='o')
plt.xlabel('Student ID')
plt.ylabel('Marks')
plt.title('Line Chart: Marks Trend')
plt.grid(True)
plt.show()
```

Histogram of Age Distribution in the class

```python
plt.hist(df['Age'], bins=5, color='lightgreen', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram: Age Distribution')
plt.show()
```

Second Bar Plot to see number of students with/without internship

```python
internship_counts = df['Internship'].value_counts()
plt.bar(internship_counts.index, internship_counts.values, color=['coral', 'lightblue'])
plt.xlabel('Internship Status')
plt.ylabel('Number of Students')
plt.title('Bar Plot: Internship Status')
plt.show()
```

Corrleation heatmap showing students with higher attendance tend to score better marks

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation
numeric_cols = df[['StudentID', 'Age', 'Marks', 'Attendance']]

# Calculate correlation matrix
corr = numeric_cols.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix - Heatmap')
plt.tight_layout()
plt.show()
```
